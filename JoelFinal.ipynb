```python
#Se crean tres joins con la finalidad de conocer la información en contexto con la informació que nos brindan otras tablas
#Join 1: Stores con Sales. Se crea vista gráfica para conocer por tienda y departamento la veta semanal, obteniendo que el departamento 2 de la tienda 12 es la que más venta semanal tiene.
join_1=storesDF.join(salesDF,storesDF.Store==salesDF.Store,"left")
display(join_1)
```

```python
#Vista del join 1:
join_1.createOrReplaceTempView("join1")
spark.sql("select * from join1").show()
```

```python
#Join 2: Stores con features:
join_2=storesDF.join(featuresDF,storesDF.Store==featuresDF.Store,"left")
display(join_2)
```

```python
#Vista del join 2:
join_2.createOrReplaceTempView("join2")
spark.sql("select * from join2").show()
```

```python
#Creación de vista obteniendo el año de la variable Date:
fechaDF.createOrReplaceTempView("fechas")
spark.sql("select *, year(fecha) from fechas").show()
```

```python
#Join 3: Sales con la base creada donde es posible obtener el año de la fecha. Se genera vista gráfica para conocer el fuel_price por fecha, obteniendo que en abril del 2012 mayor fuel_price se tiene:
join_3=salesDF.join(fechaDF,salesDF.Store==fechaDF.Store,"left")
display(join_3)
```

```python
#Vista del join 3, agregación y agrupación por año para conocer la temperatura promedio:
join_3.createOrReplaceTempView("join3")
spark.sql("select year(fecha), avg(Temperature) as prom_temp from join3 group by year(fecha) order by year(fecha)").show()
```

```python
#Obtención de agregación_1 para obtener la suma del total de las ventas del periodo analizado:
spark.sql("select sum(Weekly_Sales) as Sum_Weekly_Sales  from join1").show()
```

```python
#Obtención de agregación_2 para conocer el promedio de las ventas semanales en el periodo analizado:
spark.sql("select avg(Weekly_Sales) as Prom_Weekly_Sales  from join1").show()
```

```python
#Segunda agrupación y agregación para conocer por departamento y año la venta semanal:
spark.sql("select Dept, Date, sum(Weekly_Sales) as Sum_Weekly_Sales  from join1 group by Dept, Date").show()
```
